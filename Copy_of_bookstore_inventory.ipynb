JSON BASED INVENTORY MANAGEMENT ON BOOKSTORE


READ DATA FROM JSON

```python
import json
fd = open("records.json",'r')
r = fd.read()
fd.close()
records = json.loads(r)
```

```python
records
```

ADD NEW ITEM TO THE INVENTORY

```python
prod_id = input("Enter Book id:")
qn = int(input("Enter quantity:"))
for i in records:
  if(records[i]==prod_id):
    qn = records[prod_id]['Quantity'] + qn

name = input("Enter Book name:")
auth = input("Enter Book author:")
pubyr = int(input("Enter Publish year:"))
gnr = input("Enter genre:")
aud = input("Enter Audience:")
rtg = float(input("Enter Rating(5):"))
pr = int(input("Enter Price:"))


records[prod_id] = {'Book_Name': name, 'Book_author': auth, 'Publish_Year': pubyr, 'Genre':gnr, 'Rating':rtg,'Price':pr, 'Quantity':qn,'Audience':aud }


js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
